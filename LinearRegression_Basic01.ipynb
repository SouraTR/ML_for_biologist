# Fit a line to your data

A simple introduction to finding a best fit line for your data

**Keywords**
- Linear Regression
- `sklearn`
- Plotting

## Importing libraries

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

**Our data**

We have some spectroscopic readings where we are measuring the absorbance of light corresponding to different standard concentrations of a solution.

- We have standard concentrations in the X
- Corresponding absorbance values are in the y

After fitting the line, we can measure unknown concentrations with the help of their absorbance values 

```python
X = (0, 15, 25, 35, 45, 55)
y = (0.0002, 0.0404, 0.0679, 0.1088, 0.1483, 0.1716)
```

```python
plt.figure(figsize = (6, 6))
plt.scatter(X, y)
plt.xlabel("Concentration")
plt.ylabel("Absorbance")
```

**Reshaping our data**

We need to reshape our data to make it compatible with the functions that we are going to use

```python
X = np.array(X)
X = X.reshape(-1, 1)
y = np.array(y)
y = y.reshape(-1, 1)
```

## Create a Linear Regression Object

Now we have our data ready to perform the line fitting. We will use a `LinearRegression` object

The point to understand here is we need need to create this object and then we can use different methods inside this object on our data

```python
lr = LinearRegression()
```

```python
model = lr.fit(X, y)
```

```python
model.coef_
```

**We have fit our model to the data**

Now we can predict with new data. But we will visualise our best fit line before predicting new data

For this we will use a range of X values and predict the corresponding y values. We will plot these co-ordinates on top of our original data. This will help us visualise our model prediction line.

```python
X_fit = np.arange(X.min(), X.max(), 0.1)
X_fit = X_fit.reshape(-1, 1)
y_fit = model.predict(X_fit)
y_fit = y_fit.reshape(-1, 1)
```

```python
plt.figure(figsize = (6, 6))
plt.plot(X_fit, y_fit, color="orange")
plt.scatter(X, y, color="red")
plt.title("Best fit line")
plt.xlabel("Concentration")
plt.ylabel("Absorbance")
```

**Now lets perform some predictions**

The best fit line can work with unscaled values as well

```python
X_pred = [5, 40, 20, 50]
X_pred = np.array(X_pred).reshape(-1, 1)
y_pred = model.predict(X_pred)
```

```python
plt.figure(figsize = (6, 6))
plt.plot(X_fit, y_fit, label="best fit line")
plt.scatter(X, y, color="orange", label="training data")
plt.scatter(X_pred, y_pred, color="green", label="predicted")
plt.title("Predictions")
plt.xlabel("Concentration")
plt.ylabel("Absorbance")
plt.legend()
```
